```python
import pandas as pd
```

```python

file_path = r"C:\Users\Vagdevi\Downloads\Financial Analytics data.csv"
data = pd.read_csv(file_path)


data.head()
```

```python
# Drop the unnecessary column
data_cleaned = data.drop(columns=['Unnamed: 4'])

# Descriptive statistics for Market Capitalization and Quarterly Sales
key_metrics = data_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].describe()


data_cleaned.head(), key_metrics
```

```python
# Top 10 companies by Market Capitalization
top_market_cap = data_cleaned[['Name', 'Mar Cap - Crore']].sort_values(by='Mar Cap - Crore', ascending=False).head(10)

top_market_cap
```

```python
# Top 10 companies by Quarterly Sales
top_sales = data_cleaned[['Name', 'Sales Qtr - Crore']].sort_values(by='Sales Qtr - Crore', ascending=False).head(10)


top_sales
```

```python
# Correlation between Market Capitalization and Quarterly Sales
correlation = data_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()

correlation
```
